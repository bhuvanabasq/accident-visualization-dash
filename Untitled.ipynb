```python
# cricket

import requests
from bs4 import BeautifulSoup

res = requests.get('https://www.cricbuzz.com/cricket-series/2798/icc-mens-t20-world-cup-2021/matches')
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
total_matches = soup.findAll("a", {"class": "text-hvr-underline"})[:-2]
```

```python
temp = total_matches[0]['href'].split('/')
temp[1] = 'live-cricket-scorecard'
url = '/'.join(temp)
url
```

```python
from lxml import etree
dom = etree.HTML(str(soup))
print(dom.xpath('//*[@id="BGINP01_S1"]/section/div/font/text()'))
```
